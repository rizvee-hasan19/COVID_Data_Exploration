```python
df_sales = pd.read_csv(r"C:\Users\User\Downloads\FSDS Bootcamp\Pandas\sales_data_final.csv")
```

```python
df_sales.head()
```

```python
def profit_flag(a):
    if a >= 0:
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(-26.055)
```

```python
df_sales['profit_flag'] = df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
def len_cus(a):
    return len(a)
```

```python
len_cus('Joseph Holt')
```

```python
df_sales['len_customer'] = df_sales['customer_name'].apply(len_cus)
```

```python
df_sales.head()
```

```python
def quantity_flag(a) : 
    if  a < 5 :
        return 'low'
    elif a >5 and a <12 :
        return 'medium'
    else :
        return 'high'
```

```python
df_sales['quantity_flag'] = df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head(3)
```

```python
df_sales['square_quantity'] = df_sales['quantity'].apply(lambda a:a*a)
```
